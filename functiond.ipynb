```python
def checkMember(num):
    a = 0
    b = 1
    while b<num:
        c = a+b
        a = b
        b = c
    if b==num or a==num:
        return True
    if b > num:
        return False


n=int(input())
if(checkMember(n)):
    print("true")
else:
    print("false")
```

```python
def waveprint(li,n,m):
    for j in range(0,m,1):
        if j%2==0:
            for i in range(0,n,1):
                print(li[i][j],end=" ")
        else:
            for i in range(n-1,-1,-1):
                print(li[i][j],end=" ")
           
t=int(input())
while t>=1:
    str=input().split()
    n,m=int(str[0]),int(str[1])
    li=[]
    if m>0 and n>0:
         li=[[int(j) for j in input().split()]for i in range(n)]
    waveprint(li,n,m)
    print()
    t=t-1
```
